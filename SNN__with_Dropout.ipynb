```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
import tensorflow.keras.backend as K
import itertools

# Define custom RMSE metric
def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))

# Load and preprocess data
df = pd.read_csv("/content/drive/MyDrive/Podcast listening time Prediction using ML Algo/Podcast_Prediction.csv")

# Check and handle missing values
print("NaNs before handling:")
print(df.isnull().sum())

df['Episode_Length_minutes'].fillna(df['Episode_Length_minutes'].mean(), inplace=True)
df['Guest_Popularity_percentage'].fillna(df['Guest_Popularity_percentage'].mean(), inplace=True)

# Define features and target
X = df.drop(['Listening_Time_minutes', 'Episode_Title', 'id', 'Podcast_Name'], axis=1)
y = df['Listening_Time_minutes']

# Convert data types
numerical_cols = ['Episode_Length_minutes', 'Host_Popularity_percentage',
                 'Guest_Popularity_percentage', 'Number_of_Ads']
for col in numerical_cols:
    X[col] = X[col].astype('float32')
y = y.astype('float32')

# Define preprocessing
categorical_cols = ['Genre', 'Publication_Day', 'Publication_Time', 'Episode_Sentiment']
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
])

# Initialize KFold
kf = KFold(n_splits=3, shuffle=True, random_state=42)
histories = []
fold_results = []

# Data Visualization Before Training
plt.figure(figsize=(15, 10))

# Feature distributions
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(X[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Target vs features
plt.figure(figsize=(12, 6))
sns.scatterplot(x=X['Episode_Length_minutes'], y=y, hue=X['Genre'])
plt.title('Listening Time vs Episode Length')
plt.xlabel('Episode Length (minutes)')
plt.ylabel('Listening Time (minutes)')
plt.show()

# Cross-validation training
for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
    print(f"\nTraining Fold {fold + 1}/3")

    # Split and preprocess data
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    X_train_processed = preprocessor.fit_transform(X_train)
    X_val_processed = preprocessor.transform(X_val)

    # Build model
    model = Sequential([
        Dense(128, input_dim=X_train_processed.shape[1], activation='relu', kernel_regularizer=l2(0.001)),
        Dropout(0.3),
        Dense(64, activation='relu', kernel_regularizer=l2(0.001)),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dense(1)
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.0005),
        loss='mean_squared_error',
        metrics=[rmse]
    )

    # Train with early stopping
    early_stop = EarlyStopping(
        monitor='val_loss',
        patience=8,
        min_delta=0.001,
        restore_best_weights=True
    )

    history = model.fit(
        X_train_processed, y_train,
        validation_data=(X_val_processed, y_val),
        epochs=25,
        batch_size=32,
        callbacks=[early_stop],
        verbose=1
    )

    histories.append(history)
    fold_results.append({
        'train_loss': history.history['loss'],
        'val_loss': history.history['val_loss'],
        'train_rmse': history.history['rmse'],
        'val_rmse': history.history['val_rmse']
    })

# Plot training curves for each fold
plt.figure(figsize=(15, 10))
for i, result in enumerate(fold_results):
    plt.subplot(2, 2, 1)
    plt.plot(result['train_loss'], label=f'Fold {i+1}')
    plt.title('Training Loss')
    plt.xlabel('Epochs')
    plt.ylabel('MSE Loss')

    plt.subplot(2, 2, 2)
    plt.plot(result['val_loss'], label=f'Fold {i+1}')
    plt.title('Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('MSE Loss')

    plt.subplot(2, 2, 3)
    plt.plot(result['train_rmse'], label=f'Fold {i+1}')
    plt.title('Training RMSE')
    plt.xlabel('Epochs')
    plt.ylabel('RMSE')

    plt.subplot(2, 2, 4)
    plt.plot(result['val_rmse'], label=f'Fold {i+1}')
    plt.title('Validation RMSE')
    plt.xlabel('Epochs')
    plt.ylabel('RMSE')

for i in range(1, 5):
    plt.subplot(2, 2, i)
    plt.legend()
plt.tight_layout()
plt.show()

# Plot average training and validation loss
# Pad the sequences to the length of the longest sequence
max_len = max(len(res['train_loss']) for res in fold_results)
padded_train_loss = [list(x) for x in itertools.zip_longest(*[res['train_loss'] for res in fold_results], fillvalue=np.nan)]
padded_val_loss = [list(x) for x in itertools.zip_longest(*[res['val_loss'] for res in fold_results], fillvalue=np.nan)]
padded_train_rmse = [list(x) for x in itertools.zip_longest(*[res['train_rmse'] for res in fold_results], fillvalue=np.nan)]
padded_val_rmse = [list(x) for x in itertools.zip_longest(*[res['val_rmse'] for res in fold_results], fillvalue=np.nan)]


avg_train_loss = np.nanmean(padded_train_loss, axis=1)
avg_val_loss = np.nanmean(padded_val_loss, axis=1)
avg_train_rmse = np.nanmean(padded_train_rmse, axis=1)
avg_val_rmse = np.nanmean(padded_val_rmse, axis=1)


plt.figure(figsize=(10, 6))
plt.plot(avg_train_loss, label='Average Training Loss')
plt.plot(avg_val_loss, label='Average Validation Loss')
plt.title('Average Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(avg_train_rmse, label='Average Training RMSE')
plt.plot(avg_val_rmse, label='Average Validation RMSE')
plt.title('Average Training vs Validation RMSE')
plt.xlabel('Epochs')
plt.ylabel('RMSE')
plt.legend()
plt.show()
```
